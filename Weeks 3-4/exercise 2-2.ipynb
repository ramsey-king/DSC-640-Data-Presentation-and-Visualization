```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
world_population_df = pd.read_excel('world-population.xlsm')
world_population_df.head(7)
```

1 Line chart and 1 Step Chart

```python
# Line Chart
plt.plot(world_population_df['Year'], world_population_df['Population'])
plt.title("World Population")
plt.xlabel('Year')
plt.ylabel('Population (Billions)')
plt.show()
```

```python
# Step Chart
plt.step(world_population_df['Year'], world_population_df['Population'])
plt.title("World Population")
plt.xlabel('Year')
plt.ylabel('Population (Billions)')
plt.show()
```
